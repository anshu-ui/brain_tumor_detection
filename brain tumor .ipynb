```python
train = ImageDataGenerator(rescale=1/255)
```

```python
 train_dataset = train.flow_from_directory("D://dl dataset//brain tumor//brain_tumor_dataset",target_size=(200,200),batch_size=50,class_mode="binary")
```

```python
train_dataset.class_indices
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Conv2D(18,(3,3),activation="relu",input_shape=(200,200,3)))
model.add(MaxPool2D(2,2))
model.add(Conv2D(30,(3,3),activation="relu"))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(50,activation="relu"))
model.add(Dense(50,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
history = model.fit(train_dataset,epochs=15)
```

```python
plt.plot(history.history["accuracy"],color="darkblue")
plt.plot(history.history["loss"],color="red")
```
